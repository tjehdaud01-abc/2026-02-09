## 1. 비교 연산자 (Comparison Operators)
* **개념**: 두 값의 크기를 비교하여 참(`True`) 또는 거짓(`False`)을 결과로 반환하는 연산자입니다.
* **주요 연산자**:
    * `==` : 같다
    * `!=` : 다르다
    * `>` / `<` : 크다 / 작다
    * `>=` / `<=` : 크거나 같다 / 작거나 같다

```python
# 무작위로 정수 두개(a,b)를 생성한 다음에 a나누기b의 몫과 나머지를 출력
# a 는 10~50 사이 숫자 b 2~9사이 숫자
# a 나누기 b의 몫과 나머지
# my_quotient 몫을 예측한 값을 입력
# my_remainder 나머지를 예측한 값을 입력
# 실제 값과 예측값을 비교해서 출력 : True, False
```

```python
김지수  [오후 4:51]
import random
a = random.randint(10, 50)
b = random.randint(2, 9)

your_quoitent = a // b
your_remainder = a % b

my_quoitent = int(input('A 숫자를 입력하세요 '))
my_remainder = int(input('B 숫자를 입력하세요 '))

print(f'A 입력 결과 : {my_quoitent} || B 입력 결과 : {my_remainder}')
print(f'나눈 결과 {your_quoitent } || 나머지 {your_remainder}')
print(f'나눈 결과 {your_quoitent == my_quoitent} || 나머지 {your_remainder == my_remainder }')
```

## 2. 선택 구조 (Selection Structure)
* **개념**: 조건식의 결과에 따라 실행할 문장을 결정하는 구조입니다.
* **if 문**: 조건이 참일 때만 실행할 문장을 작성합니다.
* **if-else 문**: 조건이 참일 때와 거짓일 때 실행할 문장을 각각 나누어 작성합니다.
* **if-elif-else 문**: 여러 개의 조건을 순차적으로 체크해야 할 때 사용합니다.
* **주의 사항**:
    * 조건문 끝에는 반드시 **콜론(:)**을 붙여야 합니다.
    * 실행할 문장은 반드시 **들여쓰기(Indentation)**를 해야 합니다.

## 3. 논리 연산자 (Logical Operators)
* **개념**: 여러 개의 조건식을 조합하여 하나의 결과를 얻을 때 사용합니다.
* **종류**:
    * `and` : 두 조건이 모두 참일 때만 참
    * `or` : 두 조건 중 하나만 참이어도 참
    * `not` : 참을 거짓으로, 거짓을 참으로 반전

## 4. 함수 (Function)
* **개념**: 특정 작업을 수행하는 명령어들의 묶음입니다.
* **라이브러리 활용**: `import` 키워드를 사용하여 외부 도구(라이브러리)를 불러올 수 있습니다.
* **랜덤 함수**: `random.randint(start, end)`는 지정한 범위 내의 정수 중 하나를 무작위로 반환합니다.

## 5. [실습 예제]

### 🌡️ 온도 변환기 (온도 종류 통합형)
1. **입력**: 사용자로부터 온도 종류(섭씨 또는 화씨)와 온도 값을 입력받습니다.
2. **판단 및 계산**:
    * `if kind == '섭씨':` 이면 화씨 공식(`temp * 1.8 + 32`)으로 계산합니다.
    * `else:` 이면 화씨로 간주하고 섭씨 공식(`(temp - 32) / 1.8`)으로 계산합니다.
3. **출력**: 변환된 온도 결과를 화면에 표시합니다.



### 📐 두 도형의 넓이 비교하기
1. **계산**: 사각형(가로×세로)과 삼각형(밑변×높이/2)의 정보를 각각 입력받아 넓이를 구합니다.
2. **비교**:
    * `if 넓이1 > 넓이2:` 첫 번째 도형이 더 넓다고 출력합니다.
    * `elif 넓이1 < 넓이2:` 두 번째 도형이 더 넓다고 출력합니다.
    * `else:` 두 도형의 넓이가 같다고 출력합니다.
